```python
import numpy as np
import sklearn.datasets
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
```

```python
#low noise, plenty of samples, should be easy
X, y = sklearn.datasets.make_moons(n_samples=1000, noise=.05)
X2, y2 = sklearn.datasets.make_moons(n_samples=1000, noise=.3)
X3, y3 = sklearn.datasets.make_moons(n_samples=200, noise=.05)
X4, y4 = sklearn.datasets.make_moons(n_samples=200, noise=.3)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
plt.scatter(X2[:,0],X2[:,1],c=y2)
plt.show()
plt.scatter(X3[:,0],X3[:,1],c=y3)
plt.show()
plt.scatter(X4[:,0],X4[:,1],c=y4)
plt.axis('equal')
plt.show()
```

```python


per_clf = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf.fit(X, y)

per_clf2 = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf2.fit(X2, y2)

per_clf3 = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf3.fit(X3, y3)

per_clf4 = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf4.fit(X4, y4)

y_pred = per_clf.predict([[2, 0.5]])
y_pred2 = per_clf2.predict([[2, 0.5]])
y_pred3 = per_clf3.predict([[2, 0.5]])
y_pred4 = per_clf4.predict([[2, 0.5]])
```

# Zad.
Naucz perceptron na danych iris i wylicz F1 score.

```python
iris = load_iris()
X = iris.data[:, (2, 3)] 
y = (iris.target == 0).astype(np.int) 
```

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
per_clf = Perceptron(random_state=42)
per_clf.fit(X_train, y_train)
f1_score(y_test,per_clf.predict(X_test))
```
